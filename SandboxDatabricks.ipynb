```python
%sql
CREATE TABLE employed (
    id INT,
    name STRING,
    salary DOUBLE
)
USING DELTA;
```

```python
%sql
insert into employed
values
(1,"Adam",3500.0),
(2,"Sarah",4020.5);

insert into employed
values
(3,"jhon",2999.3),
(4,"Thomas",4000.3);

insert into employed
values
(5,"Anna",9416.3);

insert into employed
values
(5,"kim",7816.3)
```

```python
%sql
update employed
set salary = salary + 100 
where name like "A%"
```
